## Capstone Project: Predictive Health & Activity Monitoring

### What is the problem you are solving?
The core problem is the unobtrusive monitoring of human health and safety. Manual monitoring of elderly patients or high-risk workers is expensive and intrusive. By using smartphone telemetry, we can solve this by building a system that automatically distinguishes between sedentary (Sitting, Standing) and active (Walking, Climbing) behaviors. This enables automated alerts for "lack of movement" or identifying "fall risks" without the need for cameras or constant human supervision.

## What tools are you using?
I am using a modern Python-based stack to demonstrate end-to-end Machine Learning capability:  

- Data Processing: Pandas and NumPy for handling the 561-feature engineered dataset.

- Dimensionality Reduction: Scikit-Learn for PCA and KernelPCA to solve the "Curse of Dimensionality."

- Unsupervised Learning: Gaussian Mixture Models (GMM) for soft-clustering and identifying activity transition states.

- Time Series Analysis: The statsmodels and arch libraries to implement ARIMA and GARCH models for movement volatility forecasting.

Deep Learning: TensorFlow/Keras to build a high-accuracy Neural Network classifier.

## What insights do you want to discover?
I want to go beyond simple "Classification Accuracy." My goal is to discover the "Latent Structure of Movement."

The "Aha!" Moment:  
- Using GMM, I want to identify the "Probability of Transition." Instead of just saying a person is "Standing," I want the model to show when a person is in the process of sitting down (the overlap between clusters).

- Volatility Insight: Using GARCH, I want to show that we can predict "Activity Bursts." For a business, this means predicting when a user is about to transition from rest to high-intensity exercise, allowing for real-time health feedback.

How would a business or community benefit from your work?  
- Health Tech Companies: A fitness app could use this to more accurately calculate caloric burn by identifying the specific "rhythm" of walking upstairs vs. downstairs.

- Elderly Care Communities: This system could be deployed on a standard smartphone to act as a low-cost "Fall and Stagnation Detector." If the GARCH volatility drops to zero and the GMM classifies the state as "Laying" for an extended period in an unusual location, the system can trigger an emergency response.

- Efficiency: By using PCA to reduce 561 features to just 30, I am demonstrating how to save a business money on cloud computing and battery life, making the AI light enough to run locally on a phone.

## Understanding the columns
Prefix 't' vs 'f': 't' means Time (raw movement as it happened); 'f' means Frequency (the rhythm/vibration of the movement).

Body vs Gravity: Body is the person moving; Gravity is the constant pull of the earth (used to tell if you are lying down or standing up).

Acc vs Gyro: Acc (Accelerometer) measures speed and force; Gyro (Gyroscope) measures turning and rotation.

Jerk: Measures how "sudden" or "sharp" a movement was (like a quick step vs. a slow slide).

Mag (Magnitude): Combines the 3 directions (X, Y, Z) into one number showing the total "power" of the move.

X, Y, Z: The three directions in 3D space.
