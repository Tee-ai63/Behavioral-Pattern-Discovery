## Capstone Project: Predictive Health & Activity Monitoring

# A Multi-Stage Framework for Movement Discovery and Future Behavioral Forecasting

### Project Overview
This project transforms raw smartphone telemetry into actionable behavioral insights. By utilizing the UCI Human Activity Recognition (HAR) dataset, we go beyond simple classification to explore the latent structure of human movement, model movement volatility, and generate synthetic behavioral data.

The Goal: To demonstrate mastery of Unsupervised Learning (Clustering/DR), Time Series Analysis, and Deep Learning

### Dataset Structure
Data Source: Smartphone Accelerometer and Gyroscope telemetry.

Input Variables: 561 engineered features (Time and Frequency domain).

Target Variable: Activity (6 classes: Walking, Walking Upstairs, Walking Downstairs, Sitting, Standing, Laying).

Metadata: subject (ID of the person performing the activity).

## Understanding the columns
Prefix 't' vs 'f': 't' means Time (raw movement as it happened); 'f' means Frequency (the rhythm/vibration of the movement).

Body vs Gravity: Body is the person moving; Gravity is the constant pull of the earth (used to tell if you are lying down or standing up).

Acc vs Gyro: Acc (Accelerometer) measures speed and force; Gyro (Gyroscope) measures turning and rotation.

Jerk: Measures how "sudden" or "sharp" a movement was (like a quick step vs. a slow slide).

Mag (Magnitude): Combines the 3 directions (X, Y, Z) into one number showing the total "power" of the move.

X, Y, Z: The three directions in 3D space.
